```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_path = "Resources/schools_complete.csv"
student_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_path)
student_df = pd.read_csv(student_path)

# Combine the data into a single dataset.  
school_comp_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
```

```python
school_comp_df.dtypes
```

```python
school_comp_df.count()
```

```python
b_df = school_comp_df.groupby('school_name')
b_df = b_df['budget'].max()
tot_budget = b_df.sum()
```

```python
tot_schools = len(school_comp_df['school_name'].value_counts())
tot_students = school_comp_df['student_name'].count()
avg_math_score = school_comp_df['math_score'].mean()
avg_read_score = school_comp_df['reading_score'].mean()

print(f"Total Schools: {tot_schools}")
print(f"Total Students: {tot_students}")
print(f"Total Budget: {tot_budget}")
print(f"Average math score: {avg_math_score}")
print(f"Average reading score: {avg_read_score}")
```

```python
pass_math_df = school_comp_df.loc[school_comp_df["math_score"]>=70, :]
pass_read_df = school_comp_df.loc[school_comp_df["reading_score"]>=70, :]
pass_overall_df = pass_read_df.loc[pass_read_df["math_score"]>=70, :]

per_pass_math = round(len(pass_math_df["student_name"])/tot_students * 100, 2)
per_pass_read = round(len(pass_read_df["student_name"].unique())/tot_students * 100, 2)
per_pass_overall = round(len(pass_overall_df["student_name"].unique())/tot_students * 100, 2)

print(f"Percentage of students who passed math: {per_pass_math}%")
print(f"Percentage of students who passed reading: {per_pass_read}%")
print(f"Percentage of students who passed both math and reading: {per_pass_overall}%")
```

```python
district_summary = pd.DataFrame({
    "Total Schools":[tot_schools],
    "Total Students":[tot_students],
    "Total Budget":[tot_budget],
    "Average Math Score":[avg_math_score],
    "Average Reading Score": [avg_read_score],
    "% Passing Math": [per_pass_math],
    "% Passing Reading": [per_pass_read],
    "% Overall Passing": [per_pass_overall]
})
district_summary
```

```python
grp_school_df = school_comp_df.groupby(["school_name"])

school_count_df = grp_school_df['type'].value_counts()
school_count_df
```

```python
type_df = school_count_df.index.values
type_df
```

```python
school_budget = grp_school_df['budget'].max()
school_budget
```

```python
budget_per_stud = pd.merge(school_count_df, school_budget, on="school_name")
budget_per_stud['Per Student Budget'] = budget_per_stud['budget']/budget_per_stud['type']
budget_per_stud = budget_per_stud.rename(columns = {'type':"Total Students", 'budget':"Total School Budget"})
budget_per_stud
```

```python

avg_df = grp_school_df.mean()
avg_df = avg_df[['math_score', 'reading_score']]
avg_df = avg_df.rename(columns = {'math_score':"Avg Math Score", 'reading_score':'Avg Reading Score'})
avg_df
```

```python
# Need the counts to find the percentages
count_school_df = grp_school_df['Student ID'].count()
print(type(count_school_df))
count_school_df
```

```python
# % Passing Math
math_passing_df = school_comp_df.loc[school_comp_df['math_score']>=70, :]
grp_pass_math_df = math_passing_df.groupby(['school_name']).count()
grp_pass_math_df['% Passing Math'] = round(grp_pass_math_df['Student ID']/count_school_df * 100, 2)
grp_pass_math_df = grp_pass_math_df['% Passing Math']
grp_pass_math_df
```

```python
# % Passing Reading
read_passing_df = school_comp_df.loc[school_comp_df['reading_score']>=70, :]
grp_pass_read_df = read_passing_df.groupby(['school_name']).count()
grp_pass_read_df['% Passing Reading'] = round(grp_pass_read_df['Student ID']/count_school_df * 100, 2)
grp_pass_read_df = grp_pass_read_df['% Passing Reading']
grp_pass_read_df
```

```python
    # % Passing Data
passing_df = math_passing_df.loc[(school_comp_df['reading_score'] >= 70), :]
grp_pass_df = passing_df.groupby(['school_name']).count()
grp_pass_df["% Overall Passing"] = round(grp_pass_df['Student ID'] / count_school_df*100, 2)
grp_pass_df = grp_pass_df['% Overall Passing']
grp_pass_df
```

```python
# Combine % data together
grp_per_pass_df = pd.merge(grp_pass_math_df, grp_pass_read_df, on='school_name')
grp_per_pass_df = pd.merge(grp_per_pass_df, grp_pass_df, on='school_name')
grp_per_pass_df
```

```python
type_list = [t[1] for t in type_df]
type_list
```

```python
# School names
school_list = [s[0] for s in type_df]
school_list
```

```python
# Combine all summaries into a dataframe
school_type_final_df = pd.DataFrame({"school_name": school_list,
"School Type": type_list})
# "% Pass Math": math_passing_df, "% Pass Reading": read_passing_df, "% Pass Overall": grp_pass_df
school_type_final_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
s_bins = [0, 585, 630, 645, 680]
s_ranges = ['<$585', '$585-630', '$630-645', '$645-680']
s_range_df= school_summary[['Per Student Budget','Avg Math Score', 'Avg Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]
```

```python
s_range_df['Spending Ranges (Per Student)'] = pd.cut(s_range_df['Per Student Budget'], s_bins, labels=s_ranges, include_lowest=True)
sp_range_df = s_range_df.groupby('Spending Ranges (Per Student)')
avg_sp_range_df = sp_range_df.mean()
```

```python
avg_sp_range_df=avg_sp_range_df[['Avg Math Score', 'Avg Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]
avg_sp_range_df
```
